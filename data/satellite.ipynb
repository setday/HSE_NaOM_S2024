```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_json("Satellite_trajectory.json")
```

```python
n = min(len(data['current_state']), 100)

x_coord = [data['current_state'][i]['x'] for i in range(n)]
z_coord = [data['current_state'][i]['z'] for i in range(n)]
```

```python
plt.plot(x_coord, z_coord, '.', color='black')
ax = plt.gca()
ax.set_xlim([-10000, 10000])
ax.set_ylim([-10000, 10000])
```
